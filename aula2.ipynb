## Desafio

situação problema

- Joãozinho investiu no ativo da petrobrás
- R$ 100,00 (preço inicial-valor de compra): preco(i-1)
- R$ 111,00 (preço final)
- Ganho Bruto: R$ 111,00 - R$ 100,00 (R$ 11,00)
- ret = R$ 11,00/R$ 100,00 = 0.11 ganho no periodo (valorização)


```python
import numpy as np
```

```python
numpy_array_simples = df.iloc[:, 0].to_numpy()
```

```python
# mostrar quantas dimensoes o vetor tem
print(numpy_array_simples.ndim)
```

```python
# quantidade de elementos no vetor
print(numpy_array_simples.size)
```

```python
# mostre os indices do vetor
print(np.arange(numpy_array_simples.size))
```

```python
retorno = (numpy_array_simples[1:19] - numpy_array_simples[0:18]) / numpy_array_simples[0:18] 
```

```python
print(retorno)
```
