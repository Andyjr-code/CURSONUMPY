```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Definir los Datos ---
# 'tempo' (Eje X): Representa el tiempo o el índice. Usamos 20 puntos de datos.
# np.arange(0, 18.5, 0.925) crea un array que va de 0 a aproximadamente 18 con pasos.
tempo = np.linspace(0, 18, 20) 

# 'retorno' (Eje Y): Los valores que varían, simulando los retornos financieros.
# Estos datos son una aproximación a la forma de la línea de tu imagen.
retorno = np.array([
    0.023, 0.060, -0.012, 0.015, -0.012, 0.050, -0.010, 0.020, 
    -0.040, -0.055, 0.000, 0.008, 0.000, 0.008, -0.035, -0.060, 
    -0.010, 0.040, 0.030, -0.010
])

# --- 2. Crear el Gráfico de Línea ---
plt.figure(figsize=(8, 5)) # Opcional: define el tamaño

# Usamos plt.plot() para dibujar la línea
plt.plot(
    tempo,       # Datos del eje X
    retorno,     # Datos del eje Y
    color='dimgrey', # Puedes elegir el color, 'dimgrey' se parece al de la imagen
    linewidth=2      # Opcional: grosor de la línea
)

# --- 3. Agregar Etiquetas y Título (como en tu imagen) ---
plt.title('retorno', fontsize=14)
plt.xlabel('tempo', fontsize=12)
plt.ylabel('retorno', fontsize=12)

# Opcional: Configurar límites del eje Y para que se parezca más a la imagen
plt.ylim(-0.065, 0.065) 

# Opcional: Agregar una línea horizontal en Y=0 (útil para retornos)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8) 

# --- 4. Mostrar el Gráfico ---
plt.show()
```
