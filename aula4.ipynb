```python
import pandas as pd
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 8))
plt.barh(df['Nome_RA'], df['Populacao_Total'],color='purple',edgecolor='black')
plt.xlabel('População Total')
plt.show()

```

```python
plt.figure(figsize=(12, 8))

plt.bar(df['Nome_RA'], df['Populacao_Total'], color='purple', edgecolor='black')

plt.ylabel('População Total')

plt.xlabel('Nome da Região Administrativa') 

plt.xticks(rotation=45, ha='right') 

plt.show()

```

```python
plt.figure(figsize=(12, 8))
plt.plot(df['Nome_RA'], df['Populacao_Total'], marker='o', color='purple', linestyle='--')
plt.xticks(rotation=45)
plt.ylabel('População Total')
plt.title('População Total por Região Administrativa (Gráfico de Líneas)')
plt.xlabel('Nome da Região Administrativa')
plt.show()
```
